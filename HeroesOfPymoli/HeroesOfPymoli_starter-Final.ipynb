## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Filter list by gender in order to perform unique counts on each list
male_df = purchase_data[purchase_data["Gender"] == 'Male'] 
female_df = purchase_data[purchase_data["Gender"] == 'Female'] 
#trying to write conditional statement with != Male or Female. Could do by filtering by !=Male then !=Female.  Commented out below section works
nomale_df = purchase_data[purchase_data["Gender"] != "Male"]
other_df = nomale_df[nomale_df["Gender"] != "Female"]
#other_df = purchase_data[purchase_data["Gender"] == 'Other / Non-Disclosed']

#count unique names for each gender label
un_male = len(male_df["SN"].unique())
un_female = len(female_df["SN"].unique())
un_other = len(other_df["SN"].unique())

# determine percentage using unique name gender counts and total players count from above
male_per = round((un_male/player_count)*100, 2)
female_per = round((un_female/player_count)*100, 2)
other_per = round((un_other/player_count)*100, 2)
# Build DataFrame to display values
gender_df = pd.DataFrame({"Gender": ["Male", "Female", "Other / Non-Disclosed"], "Total Count": [un_male, un_female, un_other], "Percentage": [male_per, female_per, other_per]})
gender_df = gender_df.set_index(["Gender"])
gender_df['Total Count']=gender_df['Total Count']
#Format Percentage with % symbol
gender_df['Percentage']=gender_df['Percentage'].map('{:,.2f}%'.format)
gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Use groupby on Gender to get mean price and sum totals for each gender.  Pivot around gender and do computations around price
gen_mean=purchase_data.groupby('Gender', as_index=True)['Price'].mean()
gen_pp=purchase_data.groupby('Gender', as_index=True)['Price'].sum()
gen_tally=purchase_data.groupby('Gender', as_index=True)['SN'].count()

#Create data frame and populate with values.  Using gender counts from above gender-df['Total Count'] from previous dataframe
gen_agg=pd.DataFrame()
gen_agg['Mean']=round(gen_mean, 2)
gen_agg['Sum']=gen_pp
gen_agg['Avg. Purchase Total per Player'] = (gen_agg['Sum'] / gender_df['Total Count'])
gen_agg['Total Purchases'] = gen_tally

#Format data frame
gen_agg['Mean']=gen_agg['Mean'].map('${:,.2f}'.format)
gen_agg['Sum']=gen_agg['Sum'].map('${:,.2f}'.format)
gen_agg['Avg. Purchase Total per Player']=gen_agg['Avg. Purchase Total per Player'].map('${:,.2f}'.format)
gen_agg
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
## Just a quick look at the age historgram before creating bins
#max_age = purchase_data['Age'].max()
#min_age = max = purchase_data['Age'].min()
#age_check = purchase_data.groupby('Age')['Age'].count()
#age_check

#Create bins.  Went up to 999.9 to make sure including all ages in a bin.  
bins_age=[0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 44.9, 49.9]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40-44", "45+"]
#Don't want to overwrite purchase_data
age_df=pd.DataFrame(purchase_data)

#Drop duplicates from SN column.  Could be used on problems above too
age_df.drop_duplicates(['SN'], inplace=True)

#Bin Create groups/bins columns and populate using pd.cut and bin information created above
age_df['Groups']=pd.cut(age_df["Age"], bins = bins_age, labels=group_names, include_lowest = True) 
#Create duplicate value of groups so that we can groupby arond it and present .count() information.  Tried doing with same column, but had issues.
age_df['Counts']=age_df['Groups']
#reset index
age_df.reset_index()
#create new dataframe (could have just overwritten age_df instead of new dataframe to save memory)
new_age_df=pd.DataFrame(age_df.groupby('Groups')['Counts'].count())

#Add in column with percentage calculations.  Format and display
new_age_df['Percentage']=round(((new_age_df['Counts']/ player_count)*100), 2)
new_age_df['Percentage']=new_age_df['Percentage'].map('{:,.2f}%'.format)
new_age_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Don't want to overwrite purchase_data
pur_data=pd.DataFrame(purchase_data)

pur_data['Bin']=pd.cut(pur_data["Age"], bins = bins_age, labels=group_names, include_lowest = True) 
pur_data['Total Count']=pur_data['Bin']
#reset index
pur_data.reset_index()
#create new dataframe (could have just overwritten age_df instead of new dataframe to save memory)
bin_count=pd.DataFrame(pur_data.groupby('Bin')['Total Count'].count())
bin_count.reset_index()
bin_avg = round((pur_data.groupby('Bin')['Price'].mean()),2)
bin_avg
bin_sum = (pur_data.groupby('Bin')['Price'].sum())
bin_sum

bin_count['Average Purchase'] = bin_avg.map('${:,.2f}'.format)
bin_count["Total Purchase Value"] = bin_sum
bin_count['Player Counts'] = new_age_df['Counts']
bin_count['Average Total Purchase per Person'] = round(bin_count['Total Purchase Value']/bin_count['Player Counts'],2)

#Now that table is complete, will format and delte unneeded Player counts row
bin_count['Total Purchase Value']=bin_count['Total Purchase Value'].map('${:,.2f}'.format)
bin_count['Average Total Purchase per Person']=bin_count['Average Total Purchase per Person'].map('${:,.2f}'.format)
bin_count.drop('Player Counts', axis=1, inplace=True)
bin_count
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Don't want to overwrite purchase_data
tp_spend=pd.DataFrame(purchase_data)
tp_spend.reset_index()
tp_count=pd.DataFrame(tp_spend.groupby('SN')['Gender'].count())
tp_avg=round(tp_spend.groupby('SN')['Price'].mean(),2)
tp_sum=tp_spend.groupby('SN')['Price'].sum()
#Define average and format.  Define Total Purchase Value
tp_count['Average Purchase Price']=tp_avg.map('${:,.2f}'.format)
tp_count['Total Purchase Value']=tp_sum
#Rename Gender column to puchase count since gender is data i used to perform the count
tp_count.rename(columns={'Gender': 'Purchase Count'}, inplace=True)

#Sort and format values remaining unformatted values
tp_count=tp_count.sort_values('Total Purchase Value', ascending=False)
tp_count['Total Purchase Value']=tp_count['Total Purchase Value'].map('${:,.2f}'.format)
tp_count
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Don't want to overwrite purchase_data
most_pop=pd.DataFrame(purchase_data)
most_pop.reset_index()
# Writing new dataframe indexed by Item ID and Item Name.  Using the Gender field to count purchases
most_pop_count=pd.DataFrame(most_pop.groupby(['Item ID', 'Item Name'])['Gender'].count())

#Rename column Gender to Purchase Count 
most_pop_count.rename(columns={'Gender': 'Purchase Count'}, inplace=True)

#Use .sum to calculate Total Purchase Value and use .mean to calculate Item Price.  Some items have multiple prices when I first printed this.
# Multiple itmem prices for the same item id  where then averaged out using mean field.
most_totval=most_pop.groupby(['Item ID', 'Item Name'])['Price'].sum()
item_price=most_pop.groupby(['Item ID', 'Item Name'])['Price'].mean()

#Add new data to dataframe
most_pop_count['Item Price']=round(item_price, 2)
most_pop_count['Total Purchase Value']=most_totval

#Sort data
most_pop_count=most_pop_count.sort_values('Purchase Count', ascending=False)

#Format the results
most_pop_count["Total Purchase Value"]=most_pop_count["Total Purchase Value"].map('${:,.2f}'.format)
most_pop_count["Item Price"]=most_pop_count["Item Price"].map('${:,.2f}'.format)
most_pop_count.head()
```

## Most Profitable Items
